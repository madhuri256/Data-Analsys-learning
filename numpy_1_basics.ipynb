1.What is NumPy?
    NumPy (Numerical Python) is a Python library used for working with arrays. It supports large, multi-dimensional arrays and matrices, along with a collection of mathematical functions.

2.What are the advantages of NumPy over Python lists?

Faster computation using vectorization.
Less memory usage.
Support for advanced mathematical functions and broadcasting.
Easy integration with C/C++ and other libraries like Pandas, SciPy.

Random cheat sheet
        | Function                                                 | Description                                            | Example                               |
| -------------------------------------------------------- | ------------------------------------------------------ | ------------------------------------- |
| `np.random.rand(d0, d1, ...)`                            | Random floats in \[0.0, 1.0)                           | `np.random.rand(3, 2)`                |
| `np.random.randn(d0, d1, ...)`                           | Random floats from normal distribution (mean=0, std=1) | `np.random.randn(2, 2)`               |
| `np.random.randint(low, high=None, size=..., dtype=int)` | Random integers                                        | `np.random.randint(1, 10, size=(3,))` |
| `np.random.choice(a, size=..., replace=...)`             | Random sample from array                               | `np.random.choice([1, 2, 3], size=2)` |
| `np.random.uniform(low, high, size=...)`                 | Uniform distribution                                   | `np.random.uniform(10, 20, 5)`        |
| `np.random.normal(loc, scale, size=...)`                 | Normal (Gaussian) distribution                         | `np.random.normal(0, 1, 5)`           |
| `np.random.shuffle(array)`                               | Shuffles an array **in-place**                         | `np.random.shuffle(arr)`              |
| `np.random.permutation(array)`                           | Returns a shuffled copy                                | `np.random.permutation(arr)`          |
| `np.random.seed(seed)`                                   | Set seed for reproducibility                           | `np.random.seed(42)`                  |
